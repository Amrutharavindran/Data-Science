```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
a = pd.read_csv('/home/ksb/amrutha/myFile0.csv')   

# Print the dataframe
print(a)

# Drop non-numeric columns (assuming all non-numeric columns are categorical)
numeric_data = a.select_dtypes(include=[float, int])

# Print the numeric data
print(numeric_data)

# Convert to NumPy array
x = numeric_data.values

# Apply KMeans clustering
from sklearn.cluster import KMeans
kmeans = KMeans(init='k-means++', n_clusters=2, random_state=0).fit(x)
p = kmeans.predict(x)
print(p)

# Scatter plot
colors = list(map(lambda x: '#a86b32' if x == 1 else '#000000', p))
plt.scatter(x[:, 0], x[:, 1], c=colors, marker='*', picker=True)

# Show the plot
plt.show()
```
